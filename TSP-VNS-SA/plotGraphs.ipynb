```python
import numpy as np
import pandas as pd
from utils.Problem import ProblemInstance
from utils.Plots import getPoints
from utils.Problem import compute_length
from utils.Plots import plotSolution
```

```python
with open('vns_output.txt', 'r') as file:
    lines = file.readlines()
if(len(lines) < 2):
    print("generate the vns_output.txt file first!")
    exit()
problem_name = lines[0].strip()
x = list(map(int, lines[1].split()))
P = ProblemInstance(problem_name)
points = getPoints(problem_name)
for i in range(len(x)):
    x[i] -= 1
plotSolution(x, points)
found = compute_length(x, P.dist_matrix)
print("GAP IS:", P.GAP(found))
print("Solution is valid: ", P.ValidateSolution(x))
```
